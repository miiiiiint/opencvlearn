```python
import cv2
from ultralytics import YOLO
from cv2 import getTickCount, getTickFrequency
# åŠ è½½ YOLOv8 æ¨¡å‹
model = YOLO("yolov8n.pt")

# è·å–æ‘„åƒå¤´å†…å®¹ï¼Œå‚æ•° 0 è¡¨ç¤ºä½¿ç”¨é»˜è®¤çš„æ‘„åƒå¤´
cap = cv2.VideoCapture(0)

while cap.isOpened():
    loop_start = getTickCount()
    success, frame = cap.read()  # è¯»å–æ‘„åƒå¤´çš„ä¸€å¸§å›¾åƒ

    if success:
        results = model.predict(source=frame) # å¯¹å½“å‰å¸§è¿›è¡Œç›®æ ‡æ£€æµ‹å¹¶æ˜¾ç¤ºç»“æœ
    annotated_frame = results[0].plot()

    # ä¸­é—´æ”¾è‡ªå·±çš„æ˜¾ç¤ºç¨‹åº
    loop_time = getTickCount() - loop_start
    total_time = loop_time / (getTickFrequency())
    FPS = int(1 / total_time)
    # åœ¨å›¾åƒå·¦ä¸Šè§’æ·»åŠ FPSæ–‡æœ¬
    fps_text = f"FPS: {FPS:.2f}"
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 1
    font_thickness = 2
    text_color = (0, 0, 255)  # çº¢è‰²
    text_position = (10, 30)  # å·¦ä¸Šè§’ä½ç½®

    cv2.putText(annotated_frame, fps_text, text_position, font, font_scale, text_color, font_thickness)
    cv2.imshow('img', annotated_frame)
    # é€šè¿‡æŒ‰ä¸‹ 'q' é”®é€€å‡ºå¾ªç¯
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()  # é‡Šæ”¾æ‘„åƒå¤´èµ„æº
cv2.destroyAllWindows()  # å…³é—­OpenCVçª—å£
```

```python
import cv2
import numpy as np
import pyautogui
 
# è·å–å±å¹•åˆ†è¾¨ç‡
screen_width, screen_height = pyautogui.size()
 
# è®¾ç½®è§†é¢‘æ•è·çª—å£çš„å¤§å°ï¼Œå¯ä»¥æ ¹æ®éœ€è¦è¿›è¡Œè°ƒæ•´
capture_width, capture_height = 1280, 720
 
# åˆ›å»ºè§†é¢‘æ•è·å¯¹è±¡
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('screen_capture.avi', fourcc, 20.0, (capture_width, capture_height))
 
while True:
    # è·å–å±å¹•æˆªå›¾
    screenshot = pyautogui.screenshot()
 
    # å°†æˆªå›¾è½¬æ¢ä¸ºOpenCVå›¾åƒ
    frame = np.array(screenshot)
 
    # è°ƒæ•´æ•è·çª—å£çš„å¤§å°
    frame = cv2.resize(frame, (capture_width, capture_height))
 
    # å°†å±å¹•æˆªå›¾å†™å…¥è§†é¢‘æ–‡ä»¶
    out.write(frame)
 
    # æ˜¾ç¤ºå±å¹•æˆªå›¾
    cv2.imshow('Screen Capture', frame)
 
    # æŒ‰'q'é”®é€€å‡ºå¾ªç¯
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
 
# é‡Šæ”¾è§†é¢‘æ•è·å¯¹è±¡å’Œå…³é—­çª—å£
out.release()
cv2.destroyAllWindows()
 
```
